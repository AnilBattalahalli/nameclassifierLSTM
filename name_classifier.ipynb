```python
chars = [i for i in string.printable if i not in string.ascii_lowercase]
```

```python
encode = {e:i for i,e in enumerate(list(chars))}
decode = {i:e for i,e in enumerate(list(chars))}
```

```python
df = pd.read_csv("./names.csv")
df = df.fillna(value=0)

y = df[['full_name', 'first_name', 'gen_label', 'missing']].astype('int').values
```

```python
names_input = [str(i[0])+"\n"+str(i[1]) for i in df[['Person_Lname', 'Person_Fname']].values.tolist()]
```

```python
names_upper = [i.upper() for i in names_input]
```

```python
def vector_encoding(letter):
    k = len(list(encode.keys()))
    empty_vector = [0]*k
    empty_vector[encode[letter]] = 1
    return empty_vector
```

```python
def vectorize(name):
    return np.array([vector_encoding(i) for i in name])
```

```python
names_vector = [vectorize(i) for i in names_upper]
```
